### 1.1. Скачать данные по ссылке ...

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
estate = pd.read_csv('./kc-house-data.csv', header=0, encoding='latin-1')
estate.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6,4))
sns.histplot(estate['price'], bins=40)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Количество домов')
```

```python
data = estate['price']
plt.figure(figsize=(6,4))
sns.boxplot(data=[data], showfliers=False)
plt.title('Распределение стоимости недвижимости')
plt.ylabel('Цена')
```

### Выводы:
Стоимость недвижимости имеет неравномерное распределение со смещением влево. 50% значений лежит в диапазоне от 300 до 600 тысяч.<br> 95% значений лежит в диапазоне от 75 тыс до 1 млн 150 тыс. Медиана лежит в районе 450000.

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6,4))
sns.histplot(estate['sqft_living'], bins=28)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Количество домов')
```

### Выводы:
Квадратура жилой площади также имеет неравномерное распределение со смещением влево. В то же время, если не брать в анализ экстремальные значения, то в диапазоне от 200 до 4500 близко к нормальному распределению. Наибольшее количество домов (около 5000 и более) имеют площадь от 1500 до 2000.

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6,4))
sns.histplot(estate['yr_built'], bins=22)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
```

### Выводы:
Количество домов имеет тенденцию волнообразного роста с увеличением года постройки. Оновная масса домов было построено в период с конца 40-х годов 20 века по настоящее время. Самое большое количество домов имеют год постройки в период 10-е годы 21 века.

```python
plt.figure(figsize=(16, 8))
plt.scatter(estate['sqft_living'],estate['price'])
plt.xlabel('жилая площадь')
plt.ylabel('стоимость домов')
plt.title('Распределение стоимости домов в зависимости от размера жилой площади')
```

Для анализа взаимного влияния двух количественных признаков лучше подойдет диаграмма рассеяния (скаттерплот), на которой можно увидеть тенценцию роста стоимости недвижимости с увеличением площади. В то же время скученость наблюдений в зоне небольшой площади и не высокой цены демонстрирует распределение преобладание домов данной категории. <br>
На графике джоинплот видим также линию тренда, а изменив аргумент "kind=", получаем график в виде эллипсов.

```python
plt.figure(figsize=(16, 8))
sns.jointplot(x=estate['sqft_living'], y=estate['price'], kind='reg');
plt.xlabel('жилая площадь')
plt.ylabel('стоимость домов')
plt.title('Распределение стоимости домов в зависимости от размера жилой площади')
```

```python
plt.figure(figsize=(16, 8))
sns.jointplot(x=estate['sqft_living'], y=estate['price'], kind='kde');
plt.xlabel('жилая площадь')
plt.ylabel('стоимость домов')
plt.title('Распределение стоимости домов в зависимости от размера жилой площади')
```

Для визуализации взаимного влияния категориальных признаков используем боксплот и столбчатую диаграмму:

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=estate['grade'], y=estate['price'], whis=1.5)
plt.xlabel('конструкция и дизайн')
plt.ylabel('стоимость')
plt.title('Распределение стоимости домов в зависимости от конструкции и дизайна')
```

Признак 'grade' характеризует конструкцию и дизайн дома по шкале тринадцатибалльной шкале, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна. <br>
Боксплот и столбчатая диаграмма демонстрируют распределение стоимости домов в зависимоти от категории:

```python
plt.figure(figsize=(16, 8))
sns.barplot(x=estate['grade'], y=estate['price'])
plt.xlabel('конструкция и дизайн')
plt.ylabel('стоимость')
plt.title('Распределение стоимости домов в зависимости от конструкции и дизайна')
```

### Выводы:
С удучшением конструкции и дизайна дома растет ценовая категория дома.
